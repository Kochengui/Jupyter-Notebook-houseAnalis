```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('kc_house_data (2).csv')
df
```

#### 1.1. Изучите стоимости недвижимости   
    

```python
plt.figure(figsize=(10,5))
plt.hist(df['price'])
plt.xlabel('Евро')
plt.ylabel('Колличество домов')
plt.title('Стоимость недвижимости')
plt.xticks(np.arange(df['price'].min(),df['price'].max(),762377.));
# по оси х 'Евро' от 0.075(75 тысяц евро) до 7.7 ( 7.7 миллионов Евро)
plt.yticks(np.arange(0,19000,2111.111))
plt.grid();

# На графике можно заметить,что основная масса квартир (19 тысяч квартир) от 75 до 830 тысяц евро, 
# далее чуть более 2 тысяц квартир от 830 до 1 миллиона 600 тысяц евро,далее около 400 квартир 
# от 1 миллиона 600 тысяц до 3 миллиона 125 тысяч, около 100 квартир от 3 миллионов 125 тысяц до 3 миллонов 900 тысяч.

```

```python
plt.figure(figsize=(10,5))
plt.boxplot(x=df['price']);

# По выбросам можно заметить,что некоторые единицы домов внушительно отличаются от основной массы цены.

df[df['price'] > 3800000]
```

#### 1.3 Изучите распределение квадратуры жилой площади

```python
plt.figure(figsize=(10,5))
plt.hist(df['sqft_living'])
plt.xlabel('м³')
plt.ylabel('колличество')
plt.title('квадратура жилой площади')
plt.xticks(np.arange(290, 13540,1300))
plt.yticks(np.arange(0, 11000,970))
plt.grid();
# От 290 м³ до 1590 м³ около 7.7 тысяц домов, от 1590 м³ до 2900 м³ 10.67 тысяц домов, от 2900м³ до 4200 м³ около 3 тсяц домов, 
# около 500 домов от 4200м³ до 5500 м³.
```

```python
plt.figure(figsize=(10,5))
plt.boxplot(df['sqft_living']);

# По выбросам можно заметить,что некоторая жилая площадь домов внушительно отличаются от основной массы площади домов.

df[df['sqft_living']>5500]
```

#### 1.4. Изучите распределение года постройки

```python
plt.figure(figsize=(12,7))
plt.hist(df['yr_built'])
plt.xlabel('Год постройки')
plt.ylabel('Колличество')
plt.title('Год постройки домов')
plt.yticks(range(0,4000, 225))
plt.xticks(range(1900,2020,5), rotation=45)
plt.grid();
# Около 1000 домов до 1200 - 1900-1945 года постройки,
# Около 2700 до 3000 домов - 1945-2003 гожа постройки,
# C 2003 по 2015 год 3600 домов.
```

#### 2.1. Изучите распределение домов от наличия вида на набережную

```python
data_water = df['waterfront'].value_counts()
```

```python
plt.figure(figsize=(10,5))
plt.pie(data_water, labels=data_water.index, autopct='%.1f%%')
plt.title('Наличие вида на набережную');
# Не сложно понять, что 99.2 процента домов не имеют види на набережную и 0.8 процента дома с видом на набережную.
```

#### 2.2 Изучите распределение этажей домов

```python
data_floors = df['floors'].value_counts()
data_floors
```

```python
plt.figure(figsize=(15,15))
plt.pie(data_floors, labels=data_floors.index, autopct='%.2f%%')
plt.title('Этажи домов');
# 1 этажных домов 49.41%
# 2 этажных домов 38.13%
# 3 этажных 2.84%
# 1.5 этажных домов 8.84%
# 2.5 этажных домов 0.74%
# 3.5 тажных домов 0.04%
```

### 3 задача

```python
plt.figure(figsize=(8,6))
sns.heatmap(data_corr, annot=True, linewidths=0.5, cmap='coolwarm', mask=np.triu(data_corr))
plt.title('Связи между признаками');
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=df['grade'],y=df['price'])
plt.ylabel('Стоимость домов в миллионах евро')
plt.xlabel('Конструкция здания и дизайн')
plt.yticks(np.arange(0,4000000, 250000))
plt.grid()
plt.title('зависимость цены от оценки конструкции здания и дизайна');
# Не сложно увидеть, что чем выше оценка конструкции здания и дизайна, тем выше стоимость дома.
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df['sqft_living'],df['price'])
plt.xlabel('жилая площадь')
plt.ylabel('Стоимость домов в миллионах евро')
plt.title('зависимость цены от жилой площади');
# Можно увидеть сильную зависимость цены от жилой площади.
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df['view'],y=df['price']);
plt.title('завимость цены от хорошего вида');
# Отлично можно увидеть завимость цены от хорошего вида.
```
